```python
import sys
import shapefile
import matplotlib.pyplot as plt
```

# Helper functions below# 

```python
def rotate_right(lis):
    return lis[-1:] + lis[:-1]
```

```python
def average(lis):
    return 1.0 * sum(lis) / len(lis)
```

# Shape file reading #

```python
shape_path = './ClusterData/ShapeFiles/CumberlandPrecinct/'
shape_name = 'CumberlandPrecinct'

data_path = './ClusterData/ExtractedData/CumberlandPrecinct/'
centroid_data_name = 'CumberlandPrecint_CENTROIDS.txt'
```

```python
def draw_shape_file(sf, label_centroid = False, given_centroids = []):
    plt.figure(figsize = (50, 50))
    for i, shape in enumerate(sf.shapeRecords()):
        x = [i[0] for i in shape.shape.points[:]]
        y = [i[1] for i in shape.shape.points[:]]
        
        our_centroid = (average(x), average(y))
        if label_centroid:
            plt.annotate('o' + str(i), our_centroid, fontsize = 20)
        
        plt.plot(x,y)
        
    for i, centroid in enumerate(given_centroids):
        plt.annotate('g' + str(i), centroid, fontsize = 20)
    
    plt.show()
```

```python
sf = shapefile.Reader(shape_path + shape_name)
```

```python
draw_shape_file(sf, True)
```

# Shape file to OFF file converter #

```python
class OFFfile:
    def __init__(self):
        self.points = []
        self.edges = []
        self.faces = []
        self.point_indices = dict()
        
    def from_shapefile(sf):
        off_file = OFFfile()
        for shape in sf.shapeRecords():
            current_face = []
            for pt in shape.shape.points:
                index = off_file.point_indices.get(pt, len(off_file.points))
                current_face.append(index)
                if index == len(off_file.points):
                    off_file.point_indices[pt] = index
                    off_file.points.append(pt)
            for a, b in zip(current_face, rotate_right(current_face)):
                off_file.edges.append((min(a, b), max(a, b)))
            off_file.faces.append(current_face)
        off_file.edges = list(set(off_file.edges))
        return off_file
                    
    def write_to_file(self, out_file):
        v, f, e = len(self.points), len(self.faces), len(self.edges)
        print(v, f, e, file = out_file)
        for pt in self.points:
            for coordinate in pt:
                print(coordinate, end = " ", file = out_file)
            print(file = out_file)
        for face in self.faces:
            print(len(face), end = " ", file = out_file)
            for pt_index in face:
                print(pt_index, end = " ", file = out_file)
            print(file = out_file)
```
